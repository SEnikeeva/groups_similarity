```python
import pandas as pd
import numpy as np
from scipy.spatial import distance
```

constants

```python
NUM_OF_CLUSTERS = 10
MY_VEC = (-1, 1.0, 8.0, 10.0)
```

getting info about friends

```python
users = pd.read_csv('in.csv')
users = users.drop('index', 1)
users = users.sort_values(by='similarity', ascending=False)
print(users)
```

```python
def init_centers(users, NUM_OF_CLUSTERS):
    centers = []
    dst = len(users) // NUM_OF_CLUSTERS
    for i in range (NUM_OF_CLUSTERS):
        centers.append((users.loc[dst * i][1], users.loc[dst * i][2], users.loc[dst * i][3]))
    return centers
```

```python
def argmin(centers, vec):
    vector = (vec[1], vec[2], vec[3])
    dsts = []
    for i in range(NUM_OF_CLUSTERS):
        center = centers[i]
        dsts.append(distance.euclidean(center, vector))
    return np.argmin(dsts)
```

```python
def distribution(centers, users):
    clusters = []
    for i in range(NUM_OF_CLUSTERS):
        clusters.append(pd.DataFrame(columns=['id', 'similarity', 'median', 'count']))
    for i in range(len(users)):  
        index = argmin(centers, users.loc[i])
        clusters[index].loc[len(clusters[index])] = users.loc[i]
    return clusters
```

```python
def new_centers(clusters, centers):
    newcenters = []
    j = 0
    for cluster in clusters:
        summ = np.zeros(3)
        size = len(cluster)
        for i in range(size):
            vec = cluster.loc[i]
            summ += np.array([vec[1], vec[2], vec[3]])
        if size != 0:
            newcenters.append((summ / size).tolist())
        else:
            newcenters.append(centers[j])
        j += 1
    return newcenters
```

```python
def k_means(users, NUM_OF_CLUSTERS, MY_VEC):
    
    centers = init_centers(users, NUM_OF_CLUSTERS)
    flag = True
    clusters=[]
    
    while flag:
        clusters = distribution(centers, users)
        newcenters = new_centers(clusters, centers)
        flag = centers != newcenters
        centers = newcenters
        print('worked')
        
    top_cluster = clusters[argmin(centers, MY_VEC)]
    return top_cluster
```

```python
k_means(users, NUM_OF_CLUSTERS, MY_VEC)
```
